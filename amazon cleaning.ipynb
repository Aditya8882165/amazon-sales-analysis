```python
# checking sample values of discounted_price
df['discounted_price'].sample(10)
```

```python
# clean the discounted_price column 
df['discounted_price']= df['discounted_price'].str.lstrip("â‚¹")
```

```python
# replacing the symbol and change the dtype of discounted_price column
df['discounted_price'] =df['discounted_price'].str.replace(",","").astype("float")
```

```python
# checking sample values of discounted_price column again
df['discounted_price'].sample(10)
```

```python
# cleaning, replacing the symbol and change the dtype of actual_price column
df['actual_price'] =df['actual_price'].str.lstrip("â‚¹").str.replace(",","").astype("float")
```

```python
# checking sample values of discount_percentage column
df['discount_percentage'].sample(10)
```

```python
# checking sample values of rating column
df['rating'].sample(10)
```

```python
# replace the irrelevant value and change the dtype of rating column
df['rating'] =df['rating'].replace('|','0').astype(str)
```

```python
# changing dtype of rating column
df['rating']=df['rating'].astype(float)
```

```python
# checking sample values of rating column again
df['rating'].sample(10)
```

```python
# checking sample values of rating_count column
df['rating_count'].sample(10)
```

```python
# checking null values in rating_count column
df['rating_count'].isnull().sum()
```

```python
# filling nan values, changing the symbol and change the dtype of rating_count column
df['rating_count']=df['rating_count'].fillna("0").astype(str).str.replace(',','').astype('float')
```

```python
# filling the value from 0 to mean in rating_count column
df['rating_count'] =df['rating_count'].replace(0,df['rating_count'].mean()).astype(float)
```

```python
# checking sample values of rating_count colummn again
df['rating_count'].sample(10)
```

```python
# checking column names again
df.columns
```

```python
# droping the unnecessary columns
df=df.drop(['category','about_product', 'user_id', 'user_name', 'review_id'
         , 'review_title','review_content', 'img_link', 'product_link'],axis=1)
df
```

```python
# checking duplicated rows
df.duplicated().sum()
```

```python
# removing duplicate rows
df.drop_duplicates(keep='first',inplace=True)
```

```python
df.duplicated().sum()
```

```python
# cleaning the extra spacing
for i in df.columns:
    if df[i].dtype=='object':
        df[i] =df[i].str.strip()
```
